Visualizing the graph with centrality measures.  

Code found @ http://stackoverflow.com/questions/35417758/visualizing-graph-in-python-using-networkx/35708775.

```python
# Create a Spring Layout
pos=nx.spring_layout(G)
```

```python
# Find the center Node
dmin=1
ncenter=0
for n in pos:
   x,y=pos[n]
   d=(x-0.5)**2+(y-0.5)**2
   if d<dmin:
      ncenter=n
      dmin=d
    
p=nx.single_source_shortest_path_length(G,ncenter)
```

```python
p=nx.single_source_shortest_path_length(G,ncenter)
plt.figure(figsize=(8,8))
nx.draw_networkx_edges(G,pos,nodelist=[ncenter],alpha=0.4)
nx.draw_networkx_nodes(G,pos,nodelist=p.keys(),node_size=90,
   node_color=p.values(),cmap=plt.cm.Reds_r)

plt.show()
```

We see that E9 is the Central node.  Darker colors are them most central.
